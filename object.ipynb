```python
#Importing the openCV, matplot and Cvlib libraries.
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
print("Done")
```

```python
#loading the image with open OpenCV
img1 = cv2.imread(r"C:\Users\Hritik Kounsal\Desktop\1024px-Baegle_dwa.jpg")
#Image before detection
plt.imshow(img1)
plt.show()
```

```python
#using the detect function form cvlib which uses it uses 
#YOLOv3 model trained on COCO dataset capable of detecting 80 common objects in context.
bbox, label, conf = cv.detect_common_objects(img1)
```

```python
#The output image showing the result of object.
output1 = draw_bbox(img1,bbox,label,conf)
plt.imshow(output1)
plt.show()
```
